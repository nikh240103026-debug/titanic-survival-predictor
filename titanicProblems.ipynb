```python
df2['Age'].fillna(df2['Age'].median(), inplace=True)
df2['Embarked'].fillna(df2['Embarked'].mode()[0], inplace=True)
df2.drop(columns=['Cabin'], inplace = True)
df2.isnull().sum()
```

Let's start with the mosty interesting one:

```python
import matplotlib.pyplot as plt

df2[df2['Survived']==1]['Age'].hist(bins=30, alpha=0.5, color='green', label='Survived')
df2[df2['Survived']==0]['Age'].hist(bins=30, alpha=0.5, color='red', label='Did not survive')
plt.title('Age vs Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
df2[df2['Survived']==1]['Fare'].hist(bins=30, alpha=0.5, color='green', label='Survived')
df2[df2['Survived']==0]['Fare'].hist(bins=30, alpha=0.5, color='red', label='Did not survived')
plt.title('Fare vs Survival') 
plt.xlabel('Fare')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
import seaborn as sns
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df2)
plt.title('Survival Rate by Class and Gender')
plt.show()
```

```python
df2['Sex'] = df2['Sex'].map({'female':0, 'male':1})
df2['Embarked'] = df2['Embarked'].map({'S':0, 'C':1, 'Q':2})
features = ['Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch', 'Embarked']

X = df2[features]
Y = df2['Survived']

print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=200)
model.fit(X_train, Y_train)
print("Model trained successfully")
```

```python
from sklearn.metrics import accuracy_score, classification_report
Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy:  {accuracy * 100:.2f}%")
print(classification_report(Y_test, Y_pred))
```

```python
#creating new feature:
#create family size feature
df2['FamilySize'] = df2['SibSp'] + df2['Parch'] + 1 # +1 is for passenger themselves

#Create IsAlone feature
df2['IsAlone'] = (df2['FamilySize'] == 1).astype(int)
print(df2[['SibSp', 'Parch', 'FamilySize', 'IsAlone']].head(10))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Survival rate by family size
sns.barplot(x='FamilySize', y='Survived', data=df2)
plt.title('Survival rate by family size')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate alone by not alone

sns.barplot(x='IsAlone', y='Survived', data=df2)
plt.title('Survival Rate: Alone vs with family')
plt.xlabel('0=With family, 1=Alone')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Retraining the model with these new features:

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

features = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'FamilySize', 'IsAlone']

x = df2[features]
y = df2['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LogisticRegression(max_iter=200)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
print(f"New Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
```

Same accuracy. New features didn't help this model.
That's a real data science lesson — more features don't always mean better results. Sometimes they add noise instead of signal.
This is completely normal. Don't get discouraged.

So what do we do now?
Instead of adding features, let's try a different and more powerful algorithm.
Logistic Regression is simple and linear. But survival prediction is complex — it has non-linear patterns.
Let's try Random Forest — it builds multiple decision trees and combines them. Usually much better than Logistic Regression on this kind of data.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

rf_pred = rf_model.predict(x_test)
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_pred)*100:.2f}%%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

feature_importance = pd.Series(rf_model.feature_importances_, index=features)
feature_importance.sort_values().plot(kind='barh', color='steelblue')
plt.xlabel('Imporve Score')
plt.show()
```

```python
import pickle

#save the model
with open('titanic_model.pkl', 'wb') as f:
    pickle.dump(rf_model, f)

print("Model saved successfully")
```
